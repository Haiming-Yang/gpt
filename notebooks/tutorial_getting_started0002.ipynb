# Illustrative Example Revised

After observing the unnatural directions, we refine the python code so that topology matrix matches the direction of the growth. This is indicated by using 'revised' for COMPENV_MODE 

```python
L = 5
age = np.zeros(shape=(L,L))
CE = np.ones(shape=(L,L))
CE[2,2]=5
G0 = np.array([[0,0,0,0],[1,2,3,4],[2,0,2,0],[3,0,3,0],[4,0,4,0]])
GE = setEXTG0(nBSG, G0, Nj, BSG)

print('G0\n',G0)
print('GE\n',GE)

T_MAX = 3
develop(T_MAX, 
    Nj, nBSG, L, J, GE, CE, age,
    P, verbose=200, COMPENV_MODE='revised',ALPH=ALPH)
```
